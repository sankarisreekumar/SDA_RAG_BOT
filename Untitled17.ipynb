```python
import os
import chromadb
from sentence_transformers import SentenceTransformer
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
file_paths = [
    "/content/DSA.txt",
    "/content/faculty.txt",
    "/content/fee and infra.txt",
    "/content/syallbus.txt",
    "/content/elective course.txt"
]

```

```python
embed_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
chroma_client = chromadb.PersistentClient(path="./text_vectorstore")
collection = chroma_client.get_or_create_collection(name="sda_collection")
```

```python
all_chunks = []

for file_path in file_paths:
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()

        chunk_id = os.path.basename(file_path).replace(" ", "_")
        all_chunks.append((chunk_id, text))
        print(f"✔ Loaded {file_path}")
    except Exception as e:
        print(f"❌ Error loading {file_path}: {e}")
```

```python
for chunk_id, text in all_chunks:
    emb = embed_model.encode([text]).tolist()
    collection.add(
        documents=[text],
        embeddings=emb,
        ids=[chunk_id]
    )

print("\n All text files are now embedded in ChromaDB!")
```

```python
# Example: Qwen2.5-Instruct (FREE, no login)
MODEL_NAME = "Qwen/Qwen2.5-1.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    device_map="auto",
    torch_dtype=torch.float16
)

```

```python
def generate_answer(query):
    # 1️⃣ Embed the query
    q_emb = embed_model.encode([query]).tolist()

    # 2️⃣ Retrieve top chunks from Chroma
    results = collection.query(
        query_embeddings=q_emb,
        n_results=4
    )

    retrieved_text = "\n\n".join(results["documents"][0])

    # 3️⃣ Prompt for LLM
    prompt = f"""
You are an expert advisor for the School of Data Analytics (SDA).
Use ONLY the context below to answer the user's question.
If the answer is not found in the context, say 'No data found.'

### CONTEXT:
{retrieved_text}

### QUESTION:
{query}

### ANSWER:
"""

    # 4️⃣ Tokenize and generate
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    output = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.2
    )

    answer = tokenizer.decode(output[0], skip_special_tokens=True)
    return answer.split("### ANSWER:")[-1].strip()
```

```python
questions = [
    "Tell me about the course syllabus.",
    "Which electives are available?",
    "Explain the evaluation scheme.",
    "How is the course structured?",
    "How long is the project?",
    "What is the contact number of SDA?",
    "Tell me about the faculty members",
    "What is the fee structure and infrastructure?"
]

for q in questions:
    answer = generate_answer(q)
    print(f"\n❓ Question: {q}")
    print(f"💡 Answer: {answer}")
```
